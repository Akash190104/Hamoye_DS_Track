```python
X = energy[['T2']]
y = energy['T6']

lin_reg = LinearRegression()
lin_reg.fit(X, y)

pred_error = lin_reg.predict(X)
rmse = np.sqrt(mean_squared_error(y, pred_error))

print("Root Mean Squared Error (RMSE): {:.3f}".format(rmse))
```

```python
energy = energy.drop(columns=["date", "lights"])
```

```python
X = energy.drop(columns=["Appliances"])
y = energy["Appliances"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

model = LinearRegression()
model.fit(X_train, y_train)

y_train_pred = model.predict(X_train)
mae_train = mean_absolute_error(y_train, y_train_pred)

print("Mean Absolute Error (Training Set): {:.3f}".format(mae_train))
```

```python
y_train_pred = model.predict(X_train)
rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))

print("Root Mean Squared Error (Training Set): {:.3f}".format(rmse_train))
```

```python
y_test_pred = model.predict(X_test)
mae_test = mean_absolute_error(y_test, y_test_pred)
print("Mean Absolute Error (Test Set): {:.3f}".format(mae_test))
```

```python
y_test_pred = model.predict(X_test)
rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))

print("Root Mean Squared Error (Test Set): {:.3f}".format(rmse_test))
```

```python
y_train_pred = model.predict(X_train)
rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))

print("Root Mean Squared Error (Train Set): {:.3f}".format(rmse_train))
```

```python
ridge_model = Ridge()
ridge_model.fit(X_train, y_train)

y_test_pred_ridge = ridge_model.predict(X_test)
rmse_test_ridge = np.sqrt(mean_squared_error(y_test, y_test_pred_ridge))

print("RMSE on Test Set (Ridge Regression): {:.3f}".format(rmse_test_ridge))
```

```python
lasso_model = Lasso()
lasso_model.fit(X_train, y_train)

feature_weights = lasso_model.coef_
non_zero_features = sum(feature_weights != 0)

print("Number of features with non-zero weights in Lasso regression: ", non_zero_features)
```

```python
y_test_pred_lasso = lasso_model.predict(X_test)
rmse_test_lasso = np.sqrt(mean_squared_error(y_test, y_test_pred_lasso))
print("RMSE on Test Set (Lasso Regression): {:.3f}".format(rmse_test_lasso))
```
