```python
unique_Areas = food_df['Area'].value_counts()

num_unique_Areas = len(unique_Areas)
print(f"Number of unique Areas: {num_unique_Areas}")

print("Unique Areas and their counts:")
print(unique_Areas)
```

```python
unique_Items = food_df['Item'].value_counts()

num_unique_Items = len(unique_Items)
print(f"Number of unique Items: {num_unique_Items}")

print("Unique Items and their counts:")
print(unique_Items)
```

```python
unique_Elements = food_df['Element'].value_counts()

num_unique_Elements = len(unique_Elements)
print(f"Number of unique Elements: {num_unique_Elements}")

print("Unique Elements and their counts:")
print(unique_Elements)
```

I have a suspicion that the **area_code**, **item_code** and **element_code** are label encoded values of the columns *area*, *item* and *element* respectively. This means that they are NOT numeric variables but Categorical variables even if they belong to the dtype int64.

This could be easily tested out by checking the number of unique values present in these columns and see if the match with their corresponding columns

```python
num_unique_area_codes = food_df['Area Code'].nunique()
print(num_unique_area_codes)
num_unique_item_codes = food_df['Item Code'].nunique()
print(num_unique_item_codes)
num_unique_element_codes = food_df['Element Code'].nunique()
print(num_unique_element_codes)
```

Interesting. My hunch was somewhat right. The area code and element code do map but the item code is slightly different. This is good to know. Yet I think, I can conclude that it was still encoded but had a few more categories during the encoding process.
